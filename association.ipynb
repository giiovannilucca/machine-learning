```python
#importando as bibliotecas necessárias para o pipeline de ML
from mlxtend.frequent_patterns import association_rules
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
import pandas as pd
import numpy as np
```

```python
#carregando a base de dados
database = pd.read_csv('compras_mercado.csv')

#visualizando as 10 primeiras instâncias da base de dados
database.head(10)
```

```python
#filtrando as transações apenas para os itens comprados
database.fillna('1', inplace = True)

Transactions=[]

for i in range (database.shape[0]):
    transaction = []
    for j in range(database.shape[1]):
        if database.iloc[i,j] != '1':
            transaction.append(database.iloc[i,j])
    Transactions.append(transaction)

Transactions[1]
```

```python
#aplicando o one-hot encoder
te = TransactionEncoder()
te_bin = te.fit_transform(Transactions)
Transactions = pd.DataFrame(te_bin, columns = te.columns_)
```

```python
#transformando para valores numéricos
def encode(x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1
Transactions = Transactions.applymap(encode)

Transactions
```

```python
#aplicando a técnica Apriori
frequent_items = apriori(Transactions, min_support = 0.002, use_colnames = True)
frequent_items.head()
```

```python
#criando as regras de associação
rules = association_rules(frequent_items, metric='lift', min_threshold = 1)
rules.head()
```

```python
#ordenando as regras de associação pelo levantamento
rules = rules.sort_values(by='lift', ascending = False)
rules
```
